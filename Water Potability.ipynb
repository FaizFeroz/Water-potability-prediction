TRAINING THE MODEL

```python
df.describe()
```

```python
df=df.fillna(df.median())
X= df.drop('Potability',axis=1)
```

```python
Y= df['Potability']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,shuffle=True,random_state=None)
```

```python
X_train
```

```python
Y_train
```

```python
X_test
```

```python
Y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
#dt=DecisionTreeClassifier()
dt=DecisionTreeClassifier(criterion='entropy',min_samples_split=9,splitter='best')
```

```python
dt.fit(X_train,Y_train)
```

```python
prediction=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(prediction,Y_test)
```

```python
confusion_matrix(prediction,Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=1)
rf.fit(X_train,Y_train)
rf_predicn=rf.predict(X_test)
accuracy_score(Y_test,rf_predicn)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import confusion_matrix
mean_absolute_error(Y_test,rf_predicn)
```

```python
confusion_matrix(Y_test,rf_predicn)
```
